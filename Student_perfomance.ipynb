```python

import pandas as pd
import numpy as np

# Initialize the DataFrame
data = {
    'student_id': [1, 2, 3, 4],
    'name': ['anika', 'tabassum', 'farah', 'farid'],
    'subject': ['math', 'english', 'science', 'science'],
    'marks': [90, 60, 35, np.nan],
    'grades': [None, None, None, None]
}
# Create the DataFrame
df = pd.DataFrame(data)

# Assign grades based on marks
def assign_grade(marks):
    if pd.isna(marks):
      return 'NA'
    elif marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'
df['grades'] = df['marks'].apply(assign_grade)
df

```

```python
import pandas as pd

# Initialize df with sample data or an empty DataFrame
df = pd.DataFrame(columns=['student_id', 'name', 'subject', 'marks', 'grades'])

# Function to assign grades based on marks
def assign_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    else:
        return 'F'

# Add a new student
def add_student(student_id, name, subject, marks):
    global df
    new_student = {
        'student_id': student_id,
        'name': name,
        'subject': subject,
        'marks': marks,
        'grades': assign_grade(marks)
    }
    df = pd.concat([df, pd.DataFrame([new_student])], ignore_index=True)
    print(f"Student {name} added successfully!")

# Update existing student
def update_student():
    global df
    student_id = int(input("Enter Student ID to update: "))
    if student_id in df['student_id'].values:
        print("Leave blank if you do not wish to update a field.")
        name = input("Enter New Name (leave blank to keep current): ")
        subject = input("Enter New Subject (leave blank to keep current): ")
        marks = input("Enter New Marks (leave blank to keep current): ")

        if name:
            df.loc[df['student_id'] == student_id, 'name'] = name
        if subject:
            df.loc[df['student_id'] == student_id, 'subject'] = subject
        if marks:
            marks = float(marks)
            df.loc[df['student_id'] == student_id, 'marks'] = marks
            df.loc[df['student_id'] == student_id, 'grades'] = assign_grade(marks)
        print(f"Student ID {student_id} updated successfully!")
    else:
        print(f"Student ID {student_id} not found.")

# Delete a student by ID
def delete_student():
    global df
    student_id = int(input("Enter Student ID to delete: "))
    if student_id in df['student_id'].values:
        df = df[df['student_id'] != student_id].reset_index(drop=True)
        print(f"Student ID {student_id} deleted successfully!")
    else:
        print(f"Student ID {student_id} not found.")

# Display the current DataFrame
def display_students():
    print("\nCurrent Students DataFrame:")
    print(df)
df
```

```python
# 4-Class performance analysis
import numpy as np
def analyze_performance(df):
    print("\nClass Performance Analysis:")

    # Convert marks to NumPy array for faster computation
    marks = df[["Math", "English", "Science"]].to_numpy()

    # Calculate average marks for each subject
    average_marks = np.mean(marks, axis=0)
    print("\nAverage Marks Per Subject:")
    subjects = ["Math", "English", "Science"]
    for subject, avg in zip(subjects, average_marks):
        print(f"{subject}: {avg:.2f}")

    # Identify the top scorer
    total_scores = np.sum(marks, axis=1)
    top_scorer_index = np.argmax(total_scores)
    top_scorer = df.iloc[top_scorer_index]["Name"]
    print(f"\nTop Scorer: {top_scorer} (Total Marks: {total_scores[top_scorer_index]})")

    # Identify the lowest scorer
    lowest_scorer_index = np.argmin(total_scores)
    lowest_scorer = df.iloc[lowest_scorer_index]["Name"]
    print(f"Lowest Scorer: {lowest_scorer} (Total Marks: {total_scores[lowest_scorer_index]})")

    # Calculate pass percentage
    grades = df["Grade"].to_numpy()
    pass_count = np.sum(grades != "F")
    pass_percentage = (pass_count / len(grades)) * 100
    print(f"\nPass Percentage: {pass_percentage:.2f}%")
```

```python
# 6_Search for a student by ID or name
def search_student(df):
    keyword = input("Enter Student ID or Name to search: ").strip()
    results = df[(df["Student ID"].str.contains(keyword, case=False)) |
                 (df["Name"].str.contains(keyword, case=False))]

    if not results.empty:
        print("\nSearch Results:")
        print(results)
    else:
        print("No matching student found.")

# Filter students by grade
def filter_students_by_grade(df):
    grade = input("Enter grade to filter (A, B, C, F): ").strip().upper()
    if grade not in ["A", "B", "C", "F"]:
        print("Invalid grade!")
        return

    filtered = df[df["Grade"] == grade]
    if not filtered.empty:
        print(f"\nStudents with Grade {grade}:")
        print(filtered)
    else:
        print(f"No students with Grade {grade} found.")

# List students who failed any subject
def list_failed_students(df):
    failed_students = df[(df["Math"] < 40) | (df["English"] < 40) | (df["Science"] < 40)]
    if not failed_students.empty:
        print("\nStudents Who Failed in Any Subject:")
        print(failed_students)
    else:
        print("No students failed in any subject.")
```

```python
import pandas as pd

def add_student(df):
    try:
        student_id = input("Enter Student ID: ")
        if student_id in df["Student ID"].values:
            print("Student ID already exists!")
            return

        name = input("Enter Student Name: ")
        math = float(input("Enter Math Score: "))
        english = float(input("Enter English Score: "))
        science = float(input("Enter Science Score: "))

        grade = calculate_grade((math + english + science) / 3)
        df.loc[len(df)] = [student_id, name, math, english, science, grade]
        print("Student added successfully!")
    except ValueError:
        print("Invalid input. Please enter numeric values for scores.")

def update_student(df):
    student_id = input("Enter the Student ID to update: ")
    if student_id not in df["Student ID"].values:
        print("Student ID not found!")
        return

    try:
        field = input("Enter the field to update (Name, Math, English, Science): ").capitalize()
        if field not in df.columns:
            print("Invalid field!")
            return

        new_value = input(f"Enter the new value for {field}: ")
        if field in ["Math", "English", "Science"]:
            new_value = float(new_value)
        df.loc[df["Student ID"] == student_id, field] = new_value
        if field in ["Math", "English", "Science"]:
            df.loc[df["Student ID"] == student_id, "Grade"] = calculate_grade(
                df.loc[df["Student ID"] == student_id, ["Math", "English", "Science"]].mean(axis=1).values[0]
            )
        print("Student record updated successfully!")
    except ValueError:
        print("Invalid input. Please try again.")

def delete_student(df):
    student_id = input("Enter the Student ID to delete: ")
    if student_id in df["Student ID"].values:
        df.drop(df[df["Student ID"] == student_id].index, inplace=True)
        print("Student record deleted successfully!")
    else:
        print("Student ID not found!")

def analyze_performance(df):
    if df.empty:
        print("No data to analyze.")
        return

    print("\nPerformance Summary:")
    print(f"Total Students: {len(df)}")
    print(f"Average Math Score: {df['Math'].mean():.2f}")
    print(f"Average English Score: {df['English'].mean():.2f}")
    print(f"Average Science Score: {df['Science'].mean():.2f}")
    print("Grade Distribution:")
    print(df["Grade"].value_counts())

def visualize_data(df):
    if df.empty:
        print("No data to visualize.")
        return

    import matplotlib.pyplot as plt

    df["Grade"].value_counts().plot(kind="bar", title="Grade Distribution")
    plt.xlabel("Grades")
    plt.ylabel("Number of Students")
    plt.show()

def search_and_filter(df):
    query = input("Enter the name or part of the name to search: ").lower()
    filtered = df[df["Name"].str.contains(query, case=False, na=False)]
    if not filtered.empty:
        print(filtered)
    else:
        print("No records found matching the search query.")

def save_to_csv(df):
    df.to_csv("student_records.csv", index=False)
    print("Records saved to student_records.csv successfully!")

def export_filtered_data(df, grade):
    filtered = df[df["Grade"] == grade]
    if not filtered.empty:
        filtered.to_csv(f"students_with_grade_{grade}.csv", index=False)
        print(f"Filtered records exported to students_with_grade_{grade}.csv successfully!")
    else:
        print(f"No records found with Grade '{grade}'.")

def calculate_grade(avg_score):
    if avg_score >= 90:
        return "A"
    elif avg_score >= 80:
        return "B"
    elif avg_score >= 70:
        return "C"
    elif avg_score >= 60:
        return "D"
    else:
        return "F"

def main_menu():
    try:
        df = pd.read_csv("student_records.csv")
        print("Student records loaded successfully.")
    except FileNotFoundError:
        print("No existing records found. Initializing empty records.")
        df = pd.DataFrame(columns=["Student ID", "Name", "Math", "English", "Science", "Grade"])
    except pd.errors.EmptyDataError:
        print("File exists but is empty. Initializing empty records.")
        df = pd.DataFrame(columns=["Student ID", "Name", "Math", "English", "Science", "Grade"])

    while True:
        print("\nMenu:")
        print("1. Add a new student record")
        print("2. Update an existing student record")
        print("3. Delete a student record")
        print("4. View analyze_performance summary")
        print("5. Visualize data")
        print("6. Search and filter records")
        print("7. Save records to file")
        print("8. Export filtered data (e.g., Grade 'A')")
        print("9. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_student(df)
        elif choice == "2":
            update_student(df)
        elif choice == "3":
            delete_student(df)
        elif choice == "4":
            analyze_performance(df)
        elif choice == "5":
            visualize_data(df)
        elif choice == "6":
            search_and_filter(df)
        elif choice == "7":
            save_to_csv(df)
        elif choice == "8":
            grade = input("Enter the grade to filter (e.g., 'A'): ").upper()
            if grade in ["A", "B", "C", "D", "F"]:
                export_filtered_data(df, grade)
            else:
                print("Invalid grade! Please try again.")
        elif choice == "9":
            confirm = input("Are you sure you want to exit? (y/n): ").lower()
            if confirm == "y":
                print("Exiting program. Goodbye!")
                break
        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main_menu()
```
